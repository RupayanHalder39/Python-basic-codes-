###for loop

```python
x="jadavpur University"
for i in x :
  print(i)
```

```python
#take out item and print their data type
list2=[1,2,3,4.5,"jadavpur",True]
for i in list2:
  print("item ",i,"has type",type(i))
```

```python
#for complete hoar por kichu excute korbo :
for i in list2:
  print(type(i))
else:
  print("executing after for loop")
```

```python
#use of continue statement
for i in list2:
  print(type(i))
  if i==3:
    continue 
    #it sends control to for loop again so 3 is not printed
  print(i)
else:
  print("executing after for loop")
```
